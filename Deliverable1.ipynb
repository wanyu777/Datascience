```python
import pandas as pd

stud_df = pd.read_excel('student.xlsx')
```

```python
stud_df
```

```python
## Adding a Total_marks column to the data

stud_df['Total_marks'] = stud_df[['Maths', 'English', 'Physics']].sum(axis=1)
stud_df
```

```python
## Function to assign grades to students based on their total marks

def assign_grade(grade):
    if grade >= 240:
        return 'A'
    elif grade >= 200:
        return 'B'
    elif grade >= 150:
        return 'C'
    elif grade >= 120:
        return 'D'
    elif grade >= 110:
        return 'E'
    else:
        return 'F'
    
## Assigning grades to students
stud_df['Grade'] = stud_df['Total_marks'].apply(assign_grade)
stud_df
```

```python
## Filtering data to display only Males students

filter_males = stud_df[stud_df['Gender'] == 'M']
filter_males
```

```python
## Sort Male students in ascending order base on marks

male_stud_sorted = filter_males.sort_values(by = ['Total_marks'], ascending=[True])

male_stud_sorted
```

```python
## FUnction to calculate student average

def calc_average(marks):
    return marks/15

stud_df['Average'] = stud_df['Total_marks'].apply(calc_average)
stud_df
```
